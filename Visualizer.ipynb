```python
import pandas as pd
import matplotlib.pyplot as plt

df = None
last_plot = None

while True:
    print("\n========== Data Analysis & Visualization Program ==========")
    print("Please select an option:")
    print("1. Load Dataset")
    print("2. Explore Data")
    print("3. Perform DataFrame Operations")
    print("4. Handle Missing Data")
    print("5. Generate Descriptive Statistics")
    print("6. Data Visualization")
    print("7. Save Visualization")
    print("8. Exit")
    print("=====================================")

    choice = int(input("\nEnter your choice: "))

    # 1️⃣ Load Dataset
    if choice == 1:
        print("\n== Load Dataset ==")
        path = input("Enter the path of the dataset (CSV file): ")
        try:
            df = pd.read_csv(path)
            print("Dataset loaded successfully!")
        except:
            print("Error loading dataset!")

    # 2️⃣ Explore Data
    elif choice == 2:
        if df is None:
            print("Please load dataset first!")
            continue

        print("\n== Explore Data ==")
        print("1. Display the first 5 rows")
        print("2. Display the last 5 rows")
        print("3. Display column names")
        print("4. Display data types")
        print("5. Display basic info")

        sub_choice = int(input("Enter your choice: "))

        if sub_choice == 1:
            print(df.head(5))
        elif sub_choice == 2:
            print(df.tail(5))
        elif sub_choice == 3:
            print(df.columns)
        elif sub_choice == 4:
            print(df.dtypes)
        elif sub_choice == 5:
            print(df.info())
        else:
            print("Invalid choice!")

    # 3️⃣ DataFrame Operations
    elif choice == 3:
        if df is None:
            print("Load dataset first!")
            continue

        print("\n== DataFrame Operations ==")
        print("1. Sort by column")
        print("2. Filter data")

        sub_choice = int(input("Enter your choice: "))

        if sub_choice == 1:
            col = input("Enter column name to sort: ")
            print(df.sort_values(by=col))
        elif sub_choice == 2:
            col = input("Enter column name: ")
            val = input("Enter value: ")
            print(df[df[col] == val])
        else:
            print("Invalid choice!")

    # 4️⃣ Handle Missing Data
    elif choice == 4:
        if df is None:
            print("Load dataset first!")
            continue

        print("\n== Handle Missing Data ==")
        print("1. Display rows with missing values")
        print("2. Fill missing values with mean")
        print("3. Drop rows with missing values")
        print("4. Replace missing values with a specific value")

        sub_choice = int(input("Enter your choice: "))

        if df.isnull().sum().sum() == 0:
            print("No missing values found in the dataset!")
            continue

        if sub_choice == 1:
            print(df[df.isnull().any(axis=1)])

        elif sub_choice == 2:
            df.fillna(df.mean(numeric_only=True), inplace=True)
            print("Missing values filled with mean!")

        elif sub_choice == 3:
            df.dropna(inplace=True)
            print("Rows with missing values dropped!")

        elif sub_choice == 4:
            value = input("Enter value to replace: ")
            df.fillna(value, inplace=True)
            print("Missing values replaced!")

        else:
            print("Invalid choice!")

    # 5️⃣ Statistics
    elif choice == 5:
        if df is None:
            print("Load dataset first!")
            continue

        print("\n== Descriptive Statistics ==")
        print(df.describe())

    # 6️⃣ Visualization
    elif choice == 6:
        if df is None:
            print("Load dataset first!")
            continue

        print("\n== Data Visualization ==")
        print("1. Bar Plot")
        print("2. Line Plot")
        print("3. Scatter Plot")
        print("4. Pie Chart")
        print("5. Histogram")
        print("6. Stack Plot")

        sub_choice = int(input("Enter your choice: "))

        if sub_choice == 1:
            x = input("Enter x column: ")
            y = input("Enter y column: ")
            df.plot(kind='bar', x=x, y=y)
            plt.show()
            last_plot = plt

        elif sub_choice == 2:
            x = input("Enter x column: ")
            y = input("Enter y column: ")
            df.plot(kind='line', x=x, y=y)
            plt.show()
            last_plot = plt

        elif sub_choice == 3:
            print("\n== Scatter Plot ==")
            x = input("Enter x-axis column name: ")
            y = input("Enter y-axis column name: ")
            print("Generating scatter plot...")
            plt.scatter(df[x], df[y])
            plt.xlabel(x)
            plt.ylabel(y)
            plt.show()
            print("catter plot displayed successfully!")
            last_plot = plt

        elif sub_choice == 4:
            col = input("Enter column: ")
            df[col].value_counts().plot(kind='pie', startangle=90)
            plt.show()
            last_plot = plt

        elif sub_choice == 5:
            col = input("Enter column: ")
            df[col].plot(kind='hist')
            plt.show()
            last_plot = plt

        elif sub_choice == 6:
            print("Stack plot requires multiple columns")
        else:
            print("Invalid choice!")

    # 7️⃣ Save Visualization
    elif choice == 7:
        if last_plot is None:
            print("No plot to save!")
        else:
            print("\n== Save Visualization ==")
            name = input("Enter file name to save the plot (e.g., scatter_plot.png): ")
            last_plot.savefig(name)
            print(f"Visualization saved as {name} successfully!")

    # 8️⃣ Exit
    elif choice == 8:
        print("\nExiting the program. Goodbye!")
        break

    else:
        print("Invalid choice! Please try again.")
```
